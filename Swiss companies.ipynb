This small script is only made to fetch the list of swiss companies on the following wikipedia page: https://en.wikipedia.org/wiki/List_of_companies_of_Switzerland.

For this purpose, we fetch the page using requests, and then parse its content using BeautifulSoup.

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
r = requests.get('https://en.wikipedia.org/wiki/List_of_companies_of_Switzerland')
```

```python
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
content = soup.find_all(id = 'mw-content-text')
```

```python
companies = []
for x in content[0].find_all('h2'):
    for y in x.find_next_sibling('ul').find_all('a'):
        companies.append(y.string)
```

```python
print(len(companies))
companies
```

```python
with open('swiss_companies.csv', 'w') as f:
    wr = csv.writer(f, quoting=csv.QUOTE_ALL)
    wr.writerow(companies)
```
